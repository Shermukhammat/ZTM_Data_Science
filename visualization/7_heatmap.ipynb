### 4.7. Heatmap

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/museum_visitors.csv", parse_dates = [0], index_col = 0, usecols = [0, 1])
df.head(5)
```

```python
df.shape
```

```python
type(df.index)
```

```python
df['month'] = [i.month for i in df.index]
df.head(5)
```

```python
df['year'] = [i.year for i in df.index]
df.sample(5)
```

```python
df2 = df.groupby(['month', 'year']).max()
df2
```

```python
df2.unstack(level = 1)
```

```python
plt.figure(figsize = (12, 6))
xticks_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']


sns.heatmap(data = df2.unstack(level = 0), cmap = "Blues", square = True, linewidths = 1)
plt.xticks(np.arange(12), labels = xticks_labels)

plt.show()
```
