### 4.7. Heatmap

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/museum_visitors.csv", parse_dates = [0], index_col = 0, usecols = [0, 1])
df.head(5)
```

```python
df.shape
```

```python
type(df.index)
```

```python
df['month'] = [i.month for i in df.index]
df.head(5)
```

```python
df['year'] = [i.year for i in df.index]
df.sample(5)
```

```python
df2 = df.groupby(['month', 'year']).max()
df2
```

```python
df2.unstack(level = 0)
```

```python
plt.figure(figsize = (12, 6))
xticks_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']


sns.heatmap(data = df2.unstack(level = 0), cmap = "Blues", square = True, linewidths = 1)
plt.xticks(np.arange(12), labels = xticks_labels)

plt.show()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/flight_delays.csv")
df.head()
```

```python
plt.figure(figsize=(16,6))
sns.barplot(x=df['Month'], y=df['AA'])
plt.title("American Airlines kompaniyasi samolyotlari o'rtacha kechikish vaqti")
plt.ylabel("Minutes")
plt.show()
```

#### `pd.melt()` funksyasi birnechta ustunlar bitta ustunga boirlashtrish uchun ishlatiladi. `id_vars` parametri indexlar uchun `value_vars` parameti columnllar uchun

```python
pd.melt(df, id_vars = ['Month'], value_vars = ['AA', 'AS'])
```

```python
sns.catplot(data = pd.melt(df, id_vars = ['Month'], value_vars = ['AA', 'AS']),
            kind = 'bar', x = 'Month', y = 'value', hue = 'variable')

plt.show()
```

```python
sns.catplot(data = pd.melt(df, id_vars = ['Month'], value_vars = ['AA', 'AS']),
            kind = 'bar', x = 'Month', y = 'value', hue = 'variable')

plt.show()
```

```python
sns.set_color_codes("muted")
color = sns.color_palette("Reds")

plt.figure(figsize = (15, 7))
sns.heatmap(data = df.set_index('Month'), cmap = color, linewidths = 1, annot = True)
plt.title("Airline arrive time")
plt.xlabel("time")

plt.show()
```
