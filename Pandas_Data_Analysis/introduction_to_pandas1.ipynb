```python
# column from series
sets_column = pd.Series([5, 5, 5, 5, 5]) 

# New column calles seats
cars_sales["Seats"] = sets_column
cars_sales
```

```python
cars_sales['Seats'].fillna(4, inplace = True)
cars_sales
```

```python
# Column from python list
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 7.5, 9.2, 5.0, 9.6, 8.7]
cars_sales["Fuel per 100KM"] = fuel_economy
cars_sales
```

```python
cars_sales["Total fuel used (L)"] = cars_sales['Odometer (KM)'] / 100 * cars_sales["Fuel per 100KM"]
cars_sales 
```
