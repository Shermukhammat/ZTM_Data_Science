```python
def load_dataset(URL):
  return pd.read_csv(URL)
```

```python
dataset = load_dataset("https://github.com/ageron/handson-ml2/blob/master/datasets/housing/housing.csv?raw=true")
dataset.head()
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(dataset, test_size = 0.2, random_state = 31)
```

```python
housing = train_set.drop('median_house_value', axis = 1)
housing_lables = train_set['median_house_value'].copy()
```

```python
housing
```

```python
housing.info()
```

```python
# 1 Qatorlarni tashlab yuborish
housing.dropna(subset = ['total_bedrooms'])
```

```python
median = housing['total_bedrooms'].median()
median
```

```python
housing['total_bedrooms'].fillna(median, inplace = True)
```

```python
housing.info()
```

#### SimpleImputer

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy = 'median')
```

```python
housing_num = housing.drop('ocean_proximity', axis = 1)
housing_num 
```

```python
imputer.fit(housing_num)
```

```python
imputer.statistics_ # Harbir ustun uchun median qiyamat
```

```python
X = imputer.transform(housing_num)
X
```

```python
type(X)
```

```python
imputer.fit_transform(housing_num)
```

```python
housing_tr = pd.DataFrame(X, columns = housing_num.columns, index = housing_num.index)
housing_tr
```
